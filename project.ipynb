```python
#Creating a collage using open CV with minimum 8 images and add the effects like converting to gray scale, pencil sketching, shadowing, frames etc.,
from matplotlib import pyplot as plt
import cv2
import numpy as np

image1 = cv2.imread(r"D:\ai&ml\images.jpeg", 1)
image2 = cv2.imread(r"D:\ai&ml\images (7).jpeg", 1)
image3 = cv2.imread(r"D:\ai&ml\images (6).jpeg", 1)
image4 = cv2.imread(r"D:\ai&ml\images (5).jpeg", 1)
image5 = cv2.imread(r"D:\ai&ml\images (4).jpeg", 1)
image6 = cv2.imread(r"D:\ai&ml\images (3).jpeg", 1)
image7 = cv2.imread(r"D:\ai&ml\images (2).jpeg", 1)
image8 = cv2.imread(r"D:\ai&ml\images (1).jpeg", 1)
image9 = cv2.imread(r"D:\ai&ml\images (8).jpeg", 1)

image_1 = cv2.resize(image1,(200,200), interpolation = cv2.INTER_AREA)
image_2 = cv2.resize(image2,(200,200), interpolation = cv2.INTER_AREA)
image_3 = cv2.resize(image3,(200,200), interpolation = cv2.INTER_AREA)
image_4 = cv2.resize(image4,(200,200), interpolation = cv2.INTER_AREA)
image_5 = cv2.resize(image5,(200,200), interpolation = cv2.INTER_AREA)
image_6 = cv2.resize(image6,(200,200), interpolation = cv2.INTER_AREA)
image_7 = cv2.resize(image7,(200,200), interpolation = cv2.INTER_AREA)
image_8 = cv2.resize(image8,(200,200), interpolation = cv2.INTER_AREA)
image_9 = cv2.resize(image9,(200,200), interpolation = cv2.INTER_AREA)

# Now we will attach the image with another image
# We will create a horizontal stack of images
# Then we will add it to the vertical stack
# Let the horizontal pair be (image1, image2) and (image3, image4)

# Now let’s start horizontal stacking.
Horizontal1 = np.hstack([image_1,image_2,image_3])
Horizontal2 = np.hstack([image_4,image_5,image_6])
Horizontal3 = np.hstack([image_7,image_8,image_9])

# Now vertical attachment
vertical_attachment = np.vstack([Horizontal1, Horizontal2,Horizontal3])

# Show the final attachment
plt.figure(figsize=(12,12))
plt.imshow(vertical_attachment)
plt.title('Final Collage')
plt.show()
```

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np

# Loading an image
image = cv2.imread(r"D:\ai&ml\images (1).jpeg")

# Converting color to Grayscale image
img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Inverting the Gray image
inverted_gray_image = 255 - img_gray

# Now blur the image using the gaussian blur
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)

# Now inverting the blureed_img
inverted_blurred_img = 255 - blurred_img

# Now let's create the pencil sketch
pencil_sketch_img =cv2.divide(img_gray, inverted_blurred_img, scale = 250.0)

plt.figure(figsize=(10,10))
plt.imshow(pencil_sketch_img, cmap = "gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the input image
input_image = cv2.imread(r'D:\ai&ml\images (1).jpeg')

# Define a shadowing kernel
kernel = np.array([[0, 0, 0], [1, 0, -1], [0, 0, 0]])

 # Apply the shadow effect
shadowed_image = cv2.filter2D(input_image, -1, kernel)

 # Display shadowed image using matplotlib
plt.figure(figsize=(10, 10))

# Display shadowed image
plt.imshow(cv2.cvtColor(shadowed_image, cv2.COLOR_BGR2RGB))
plt.title('Shadowed Image')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
# Define frame parameters
color = (0, 255, 0)  # Green color (BGR format)
thickness = 30       # Frame thickness

 # Add frame to the image
height, width = input_image.shape[:2]
framed_image = cv2.rectangle(input_image, (0, 0), (width - 1, height - 1), color, thickness)

 # Display original and framed image using matplotlib
plt.figure(figsize=(10,10))
# Display framed image
plt.imshow(cv2.cvtColor(framed_image, cv2.COLOR_BGR2RGB))
plt.title('Framed Image')
plt.axis('off')
plt.show()
```
